```python
#Question 1 - Turning Tables

rows = int(input("Enter number of rows: "))
print("<table>")

for i in range(1,rows+1):
    if i%2==0:
        print('<tr class="white row"><td>',i,'</td><td>',i*i,'</td></tr>',sep="")
    else:
        print('<tr class="lightyellowrow"><td>',i,'</td><td>',i*i,'</td></tr>',sep="")

print("</table>")
```

```python
#Question 3 - Followers

def obedient(number):
    if number%3 == 0:
        middle = number / 3
        lower = middle - 1
        upper = middle + 1
        if middle + lower + upper == number:
            return True
    else:
        return False
    
obedient(123)
```

```python
#Question 4 - Go Forth and Multiply

def multtable(size):
    for row in range(1,size+1):
        for columns in range(1,size+1):
            print(row*columns, end="\t")
        print()

multtable(4)
```
