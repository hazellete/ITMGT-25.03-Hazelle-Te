```python
#write code here 

import re

#Defined a function to fix the years 
def fix_year(birthyear):
    birthyear=str(birthyear)
    suffix = re.findall(r"[0-9][0-9]$",birthyear) #Get the last 2 digits of the birthyear using RegEx
    
    if int(suffix[0]) < 2 and int(suffix[0]) >= 0:
        year = "20" + suffix[0]  #Added '20' before the suffix if it is within the range given
    elif int(suffix[0]) >= 97 and int(suffix[0]) < 100:
        year = "19" + suffix[0]  #Added '19' before the suffix if it is within the range given
    return year

recweek['birthyear'] = recweek['birthyear'].apply(fix_year)
recweek
```

```python
#Created a dictionary containing possible dirty course names as keys and the correct course names as values
course_dict = {"BS ITE":"BS IT Entrepreneurship","ITE":"BS IT Entrepreneurship","Information Technology Entrepreneurship":"BS IT Entrepreneurship",
               "BS ME":"BS Management Engineering","ME":"BS Management Engineering","Management Engineering":"BS Management Engineering",
               "BS MGT":"BS Management","Management":"BS Management","MGT":"BS Management",
               "BS LM":"BS Legal Management","LM":"BS Legal Management","Legal Management":"BS Legal Management",
               "BS CTM":"BS Communications Technology","CTM":"BS Communications Technology","Communications Technology":"BS Communications Technology"
               }

#Defined a function to fix the course names by accessing the dictionary
def fix_courses(course):
    for key in course_dict.keys():
        if course == key:
            return course_dict[key]
        else:
            pass
        
recweek['course'] = recweek['course'].apply(fix_courses)
recweek
```
